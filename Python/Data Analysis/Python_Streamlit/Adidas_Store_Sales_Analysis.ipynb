```python
category_df = df.groupby(by = ["Product"], as_index = False)["TotalSales"].sum()
```

```python
category_df
```

```python
fig = px.bar(category_df, x = "Product", y = "TotalSales", 
             text = ['${:,.2f}'.format(x) for x in category_df["TotalSales"]],
             template = "seaborn")

fig.show()
```

```python
fig = px.pie(df, values = "TotalSales", names = "Region", hole = 0.5)
fig.update_traces(text = df["Region"], textposition = "outside")
fig.show()
```

```python
df["month_year"] = df["InvoiceDate"].dt.to_period("M")
```

```python
df
```

```python
linechart = pd.DataFrame(df.groupby(df["month_year"].dt.strftime("%Y : %b"))["TotalSales"].sum()).reset_index()
fig2 = px.line(linechart, x = "month_year", y="TotalSales", labels = {"Sales": "Amount"},height=500, width = 1000,template="gridon")
fig2.show()
```

```python
fig = px.pie(df, values = "TotalSales", names = "SalesMethod", template = "plotly_dark")
fig.update_traces(text = df["SalesMethod"], textposition = "inside")
```

```python
# Create a scatter plot
# ['Retailer', 'RetailerID', 'InvoiceDate', 'Region', 'State', 'City','Product', 'PriceperUnit', 'UnitsSold', 'TotalSales', 'OperatingProfit','OperatingMargin', 'SalesMethod']
data1 = px.scatter(df, x = "TotalSales", y = "OperatingProfit", size = "UnitsSold")
data1['layout'].update(title="Relationship between Sales and Profits using Scatter Plot",
                       titlefont = dict(size=15),xaxis = dict(title="Sales",titlefont=dict(size=19)),
                       yaxis = dict(title = "Profit", titlefont = dict(size=19)))

data1.show()
```

```python
fig = px.pie(df, values = "TotalSales", names = "Product", template = "gridon")
fig.update_traces(text = df["Product"], textposition = "inside")
fig.show()
```
