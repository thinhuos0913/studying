```python
import pandas as pd
import glob

df1 = pd.read_csv('comment.csv')
df2 = pd.read_csv('scraped_imdb.csv')
df3 = pd.read_csv('scraped_pdb.csv')
```

```python
# Get a list of all CSV files in a directory
path = "D:\JOBS\_PortfolioProjects\studying\Python\Data Processing\Crawl data\*"
csv_files = glob.glob(path + '.csv')

# Create an empty dataframe to store the combined data
combined_df = pd.DataFrame()

# Loop through each CSV file and append its contents to the combined dataframe
for csv_file in csv_files:
    df = pd.read_csv(csv_file) # not use header and index => set (index_col=None, header=0)
    combined_df = pd.concat([combined_df, df])

# Print the combined dataframe
print(combined_df)
```
