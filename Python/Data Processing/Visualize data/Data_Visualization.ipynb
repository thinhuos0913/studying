```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
filepath = "insurance.csv"
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
sns.scatterplot(x=data['bmi'], y=data['charges'])
# plt.show()
```

```python
sns.regplot(x=data['bmi'], y=data['charges'])
```

```python
sns.scatterplot(x=data['bmi'], y=data['charges'], hue=data['smoker'])
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=data)
```

```python
sns.swarmplot(x=data['smoker'],
              y=data['charges'])
```

```python
sns.lineplot(data=data['FTHG'], label="FTHG")
sns.lineplot(data=data['FTAG'], label="FTAG")
# plt.figure(figsize=(10,6))
plt.xlabel("NUMB OF MATCH")
plt.ylabel("GOAL")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("FT HOME GOAL OF SEASON")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=data.index, y=data['FTHG'])

# Add label for vertical axis
plt.ylabel("GOAL")
plt.show()
```
